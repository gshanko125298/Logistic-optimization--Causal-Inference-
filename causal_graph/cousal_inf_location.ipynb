```python
from causalgraphicalmodels import CausalGraphicalModel
```

```python
sprinkler = CausalGraphicalModel(
    nodes=["Location","Distance", "lessthan 5KM", "between (5-10)kM", "shortest distance(10-20)KM", "logest distance(grater than 25KM)", "Driver_Accept","Driver_Reject","Fulfilled_order","Unfulfilled_order"],
    edges=[
        ("Location", "Distance"), 
        ("Distance", "lessthan 5KM"), 
        ("Distance", "between (5-10)kM"),
        ("Distance", "shortest distance(10-20)KM"), 
        ("Distance", "logest distance(grater than 25KM)"),
        ("lessthan 5KM","Driver_Accept"),
        ("between (5-10)kM","Driver_Accept"),
        ("shortest distance(10-20)KM","Driver_Accept"),
        ("logest distance(grater than 25KM)","Driver_Reject"),
        ("Driver_Accept","Fulfilled_order"),
        ("Driver_Reject","Unfulfilled_order")
               
        
    ]
)
# draw return a graphviz `dot` object, which jupyter can render
sprinkler.draw()
```

```python
# get the distribution implied by the graph
print(sprinkler.get_distribution())
```
